```python
import requests 
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
url    =  'https://www.worldometers.info/coronavirus/'
web_url=   requests.get(url).content
soup   =   BeautifulSoup(web_url,'html.parser')
```

```python

data=[]
table = soup.find('table',{"id":"main_table_countries_today"})
```

```python
titles=[col.text for col in table.findAll('th')]
titles[0] = 'Country'


values = table.find('tbody').findAll({'tr':'row'})

for entry in values:
    entries=[x.text for x in entry.findAll('td')]
    data.append(entries)
    
df= pd.DataFrame(data,columns=titles)

df
```
